## Histogram

{gia tri: so lan xuat hien}

```python
arr = np.array([10, 20, 30, 10, 10, 20])
```

```python
def calculate_histogram(arr):
    
    # O(n^2)
    
    hist = {}
    
    for i in set(arr):
        
        count = 0
        
        for j in arr:
            if j == i:
                count += 1 
        
        hist[i] = count 
    
    return hist
```

```python
calculate_histogram(arr)
```

```python
def calculate_histogram(arr):
    
    # O(n)
    
    hist = {}
    
    for i in range(len(arr)):
         
        if arr[i] in hist:
            hist[arr[i]] += 1
        else:
            hist[arr[i]] = 1
            
    return hist
```

```python
def calculate_histogram(arr):
    """
    Function to calculate histogram of an array
    
    * Parameters:
    arr -- an numpy array, with shape (n,)
    
    * Return:
    hist --  a dictionary, indicate result histogram
    
    1. 
    2. 
    3. 
    """
    
    hist = {}
    
    for i in range(len(arr)):
        hist[arr[i]] = hist.get(arr[i], 0) + 1
            
    return hist
```

```python
hist = calculate_histogram(arr)
```

```python
plt.xticks(ticks=[10,15, 20,25, 30])
plt.yticks(ticks=[1, 2, 3])
plt.bar(x = hist.keys(), height = hist.values())
```

## Nhiều biểu đồ trên cùng 1 ảnh

## plt.text, plt.Rectangle, plt.Circle, plt.Polygon

## Bar chart chồng lên nhau

```python
arr_1 = np.array([10, 20, 60, 35, 75, 100, 65])
arr_2 = np.array([10, 20, 60, 35, 75, 100, 65])

plt.bar(x=np.arange(len(arr_1)), height=arr_1, color='red', label='số 1')
plt.bar(x=np.arange(len(arr_2)), height=arr_2, color='green', bottom=arr_1, label='số 2')
plt.legend()
```

## Bar chart liền kề nhau

```python
# x = [0, 1, 2, 3, 4, 5, 6]
# x = [0, 1, 2, 3, 4, 5, 6] + 0.2

WIDTH = 0.3

arr_1 = np.array([100, 20, 60, 350, 75, 100, 65])
arr_2 = np.array([10, 20, 60, 35, 75, 100, 65])
arr_3 = np.array([10, 20, 60, 35, 75, 100, 65])

plt.bar(x=np.arange(len(arr_1)), height=arr_1, color='red', label='số 1', width=WIDTH)
plt.bar(x=np.arange(len(arr_2)) + WIDTH, height=arr_2, color='green', label='số 2', width=WIDTH)
plt.bar(x=np.arange(len(arr_3)) + 2*WIDTH, height=arr_3, color='blue', label='số 3', width=WIDTH)
plt.legend()
```
