Ex. Given a following data frame:

data = pd.DataFrame({
'Qu1': [1, 3, 4, 3, 4],

'Qu2': [2, 3, 1, 2, 3],

'Qu3': [1, 5, 2, 4, 4]})

Compute a histogram on multiple related columns.

Expected result:
image.png


```python
data = pd.DataFrame({'Qu1': [1, 3, 4, 3, 4],
                     'Qu2': [2, 3, 1, 2, 3],
                     'Qu3': [1, 5, 2, 4, 4]})

result = data.apply(pd.value_counts).fillna(0)
result
```
